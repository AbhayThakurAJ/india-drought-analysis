```python
import pandas as pd
import matplotlib.pyplot as plt

# Load rainfall data
rain = pd.read_excel(
    "/kaggle/input/india-drought-analysis-data-2000-2023/climate_data/chirps_rainfall_timeseries.xlsx"
)

# Convert date to datetime and extract year
rain["date"] = pd.to_datetime(rain["date"])
rain["Year"] = rain["date"].dt.year

# Plot yearly average rainfall
rain.groupby("Year")["precipitation"].mean().plot(figsize=(8,4))

plt.title("Average Annual Rainfall Trend (CHIRPS)")
plt.xlabel("Year")
plt.ylabel("Rainfall (mm)")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load SPEI data
spei = pd.read_excel(
    "/kaggle/input/india-drought-analysis-data-2000-2023/climate_data/marathwada_spei.xlsx"
)

# Convert date column and extract year
spei["dates"] = pd.to_datetime(spei["dates"])
spei["Year"] = spei["dates"].dt.year

# Plot yearly average SPEI
spei.groupby("Year")["spei"].mean().plot(figsize=(8,4))
plt.axhline(0, color="red", linestyle="--")

plt.title("Drought Severity Trend using SPEI Index")
plt.xlabel("Year")
plt.ylabel("SPEI Value")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
grace = pd.read_csv(
    "/kaggle/input/india-drought-analysis-data-2000-2023/groundwater_data/grace_2009_2017.csv"
)

# Convert date to datetime and extract year
grace["date"] = pd.to_datetime(grace["date"])
grace["Year"] = grace["date"].dt.year

# Plot yearly average groundwater anomaly
grace.groupby("Year")["mean"].mean().plot(figsize=(8,4))

plt.title("Groundwater Anomaly Trend from GRACE Satellite (2009â€“2017)")
plt.xlabel("Year")
plt.ylabel("Groundwater Anomaly")
plt.grid(True)
plt.show()
```
